```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GroupShuffleSplit
```

```python
from sklearn.metrics import make_scorer

def loss(y_true, y_pred):
    
    if isinstance(y_true, pd.Series):
        y_true = y_true.values

    true = np.maximum(5., np.log10(np.maximum(1., y_true)))
    pred = np.maximum(5., np.log10(np.maximum(1., y_pred)))
    
    loss = np.mean(np.abs(true - pred))
    
    return loss
    
fan_loss = make_scorer(loss, greater_is_better=False)
```

```python
def process_date(X):
    date = pd.to_datetime(X['Fiscal_year_end_date'], format='%Y-%m-%d')
    return np.c_[date.dt.year, date.dt.month, date.dt.day]

date_transformer = FunctionTransformer(process_date, validate=False)
```

```python
def process_APE(X):
    APE = X['Activity_code (APE)'].str[:2]
    APE2 = X['Activity_code (APE)'].str[:4]
    #APE3 = X['Activity_code (APE)'].str[2:4]
    return np.c_[pd.to_numeric(APE, errors='coerce').values,
                 pd.to_numeric(APE2, errors='coerce').values,
                 #pd.to_numeric(APE3, errors='coerce').values
                ]

APE_transformer = FunctionTransformer(process_APE, validate=False)
```

```python
award_cols = ['Name_processed',
              #'End_of_call_year'
             ]
X_award_cols = ['Name_processed',
              #    'Year'
                 ]
award_att = [
            #'mean',
            'count',
           'max',
           #'min',
           #'sum'
           ]
bid_att = [
            #'mean',
            'count',
           'max',
           #'min',
           #'sum'
           ]
award_features = award.groupby(award_cols)['amount'].agg(award_att)
bids_features = award.groupby(award_cols)['number_of_received_bids'].agg(bid_att)


def award_merge_naive(X):
    df1 = pd.merge(X, award_features, left_on=X_award_cols, 
                  right_on=award_cols, how='left')
    df2 = pd.merge(X, bids_features, left_on=X_award_cols, 
                  right_on=award_cols, how='left')
    return np.c_[df1[award_att],df1[award_att]]
award_merge_transformer = FunctionTransformer(award_merge_naive, validate=False)
```

```python
numeric_transformer = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='mean'))
])
```

```python
zipcode_col = ['Zipcode']
date_cols = ['Fiscal_year_end_date']
drop_cols = ['Name', 'Address', 'City']
APE_col = ['Activity_code (APE)']
num_cols = [
            'Legal_ID', 
            'Headcount', 
            #'Fiscal_year_duration_in_months', 
            'Year'
            ]
cat_cols = ['Activity_code (APE)',
            #'Name',
            #'Address',
            #'City',
            #'Last_APE'
            ]
merge_col = ['Legal_ID']


preprocessor_comp = ColumnTransformer(
    transformers=[
        #('zipcode', make_pipeline(zipcode_transformer, SimpleImputer(strategy='median')), zipcode_col),
        ('num', numeric_transformer, num_cols),
        #('date', make_pipeline(date_transformer, SimpleImputer(strategy='median')), date_cols),
        ('APE', make_pipeline(APE_transformer, SimpleImputer(strategy='mean')), APE_col),
        ('merge headcounts', make_pipeline(merge_transformer, SimpleImputer(strategy='mean')), head_cols),
        ('merge awards', make_pipeline(award_merge_transformer, SimpleImputer(strategy='mean')), X_award_cols),
        #('category', make_pipeline(category_transformer, SimpleImputer(strategy='median')), cat_cols)
        #('drop cols', 'drop', drop_cols),
    ])
```

```python
clf = Pipeline(steps=[
    ('preprocessing', preprocessor_comp),
    ('classifier', regressor)])

cv = GroupShuffleSplit(n_splits=8, test_size=0.25)

#n_samples = 100000
n_samples = len(X_df)
scores_Xdf = -cross_val_score(clf, X_df[:n_samples], y_array[:n_samples], cv=cv, scoring=fan_loss, groups=X_df['Legal_ID'][:n_samples], n_jobs=2)

print("mean: %e (+/- %e)" % (scores_Xdf.mean(), scores_Xdf.std()))
```
